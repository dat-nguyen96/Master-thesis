```python
import feedparser
import pandas as pd
```

```python
d = feedparser.parse('http://feeds.bbci.co.uk/afrique/sports/rss.xml')
```

```python
data = pd.DataFrame([])
for post in d.entries:
    try:
        title = post.title
        link = post.link
        date = post.published
        data = data.append(pd.DataFrame({'title': title, 'url': link,'date' : date}, index=[0]), ignore_index=True)
    except:
        continue
```
